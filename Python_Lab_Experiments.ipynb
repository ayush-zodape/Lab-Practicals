### 1.4: Area of Triangle using Heron's Formula

```python
import math

def triangle_area_heron(a, b, c):
    s = (a + b + c) / 2
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

a = float(input("Enter the first side of the triangle: "))
b = float(input("Enter the second side of the triangle: "))
c = float(input("Enter the third side of the triangle: "))
print("The area of triangle is", triangle_area_heron(a, b, c))
```

### 1.5: Area of Rectangle using Function

```python
import math

def rectangle_area(l, b):
    area = l * b
    return area

l = float(input("Enter the length of the Rectangle: "))
b = float(input("Enter the breadth of the Rectangle: "))
print("The area of rectangle is", rectangle_area(l, b))
```

### 1.6: Area of Circle using Function

```python
import math

def circle_area(r, pi):
    area = pi * r * r
    return area

pi = 22/7
r = float(input("Enter the radius of the circle: "))
print("The area of circle is", circle_area(r, pi))
```

## Experiment 4: Remove i-th Occurrence of Word in List

**Date:** 01/09/2025

### 4.1: Delete i-th Occurrence Using del Function

```python
# Step 1: Declare a list with repeating words
word_list = ["apple", "banana", "cherry", "apple", "orange", "apple", "banana"]

# Step 2: Ask the user for the word and the occurrence index to delete
word_to_delete = input("Enter the word you want to delete: ")
occurrence_index = int(input("Enter the occurrence number (starting from 1) to delete: "))

# Step 3: Find and delete the ith occurrence
new_list = word_list[:] # Make a copy of the original list to store the result
occurrence_count = 0 # To track the occurrences of the word

for index in range(len(new_list)):
    if new_list[index] == word_to_delete:
        occurrence_count += 1
        if occurrence_count == occurrence_index:
            del new_list[index]
            break # Exit the loop after deleting the ith occurrence

# Step 4: Print the new list
print("Updated list:", new_list)
```

### 4.2: Remove Every Even Occurrence

```python
# Step 1: Declare a list with repeating words
word_list = ["apple", "banana", "apple", "orange", "apple", "banana", "apple"]

# Step 2: Ask the user for the word to remove even occurrences
word_to_remove = input("Enter the word you want to remove every even occurrence of: ")

# Step 3: Track the occurrences and modify the list
new_list = [] # To store the new list after removing even occurrences
occurrence_count = 0 # To keep track of the number of occurrences of the word

# Loop through the list and process each word
for word in word_list:
    if word == word_to_remove:
        occurrence_count += 1
        if occurrence_count % 2 == 0: # Check if the occurrence is even
            continue # Skip adding this word to the new list
    new_list.append(word)

# Step 4: Print the new list
print("Updated list:", new_list)
```

### 5.3: Find Most and Least Frequent Words

```python
def count_occurrence(string, w):
    words = string.split()
    
    max_count_word = words[0]
    min_count_word = words[0]
    
    for word in words:
        if words.count(word) > words.count(max_count_word):
            max_count_word = word
        if words.count(word) < words.count(min_count_word):
            min_count_word = word
    
    print("Occurrence of word", w, "is", words.count(w))
    print("Word that occurred most in the string is:", max_count_word)
    print("Word that occurred least in the string is:", min_count_word)

string = input("Enter the string: ")
w = input("Enter the word to count the number of occurrence: ")
count_occurrence(string, w)
```

## Experiment 8: Count Frequency Using Dictionary

**Date:** 31/10/2025

### 8.1: Word Frequency Using zip()

```python
def word_frequency(text):
    # Split the string into words
    words = text.split()
    
    # Create a dictionary to store the word frequencies
    word_count = {}
    
    # Use zip() to pair words with their count using a dictionary comprehension
    for word, count in zip(words, [words.count(word) for word in words]):
        word_count[word] = word_count.get(word, 0) + count
    
    # Remove duplicates by converting to a dictionary of unique keys
    return word_count

# Example usage
input_string = "hello world hello everyone world hello"
result = word_frequency(input_string)
print(result)
```

### 8.2: Dictionary of Numbers and Their Squares

## Experiment 9: Nested Dictionary with First Character as Key

**Date:** 31/10/2025

### 9.1: Nested Dictionary with Student Records

```python
students = {}

# Number of students
n = int(input("Enter the number of students: "))

# Collect student details
for i in range(1, n + 1):
    name = input(f"Enter the name of student {i}: ")
    age = int(input(f"Enter the age of student {i}: "))
    grade = input(f"Enter the grade of student {i}: ")
    
    # Create a dictionary for each student
    students[i] = {
        'Name': name,
        'Age': age,
        'Grade': grade
    }

# Display the nested dictionary
print("\nStudent Records:")
for serial, record in students.items():
    print(f"Serial No: {serial} => {record}")
```

### 9.2: Nested Dictionary with First Letter and Word Frequency

```python
from collections import defaultdict

# Input string
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Create a nested dictionary where the first letter of the word is the key
nested_dict = defaultdict(lambda: defaultdict(int))

# Process each word
for word in words:
    first_letter = word[0].lower() # Get the first letter (case-insensitive)
    nested_dict[first_letter][word.lower()] += 1 # Increment the frequency of the word

# Display the nested dictionary
print("\nNested Dictionary (First letter -> Word -> Frequency):")
for letter, word_freq in nested_dict.items():
    print(f"Letter: {letter.upper()}")
    for word, freq in word_freq.items():
        print(f"  Word: '{word}', Frequency: {freq}")
```

## Experiment 10: Recursion Programs

**Date:** 31/10/2025

### 10.1: Flatten Nested List Using Recursion

```python
def flatten(nested_list):
    flat_list = []
    
    # Loop through each item in the nested list
    for item in nested_list:
        # If item is a list, recursively flatten it
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            # If item is not a list, add it to the flat list
            flat_list.append(item)
    
    return flat_list

# Example usage
nested_list = [1, [2, 3], [4, [5, 6]], 7]
flattened_list = flatten(nested_list)
print("Flattened List:", flattened_list)
```

### 10.2: Sum of Elements and Remainders Using Recursion

```python
def sum_elements(lst):
    if not lst:
        return 0
    return lst[0] + sum_elements(lst[1:])

def sum_remainders(lst):
    if not lst:
        return 0
    return (lst[0] % 2) + sum_remainders(lst[1:])

# Example usage:
lst = [1, 2, 3, 4, 5]
sum_of_elements = sum_elements(lst)
sum_of_remainders = sum_remainders(lst)

print("Sum of elements:", sum_of_elements)
print("Sum of remainders when divided by 2:", sum_of_remainders)
```
